```python
df_bin = pd.DataFrame( data ={"Bin":["18-29","30-44", "45-59", "60+"]})
df_bin.head()
```

```python
def impute_age_bin(cols):
    bin_18 = bin_30= bin_45= bin_60= 0
    if cols >= 18 and cols<30:
        bin_18 += bin_18
    elif cols >= 30 and cols<45:
        bin_30 += bin_30
    elif cols >= 45 and cols<60:
        bin_45 += bin_45
    else:
        bin_60 += bin_60
    
    return pd.DataFrame({"percent": map(lambda x: x*100/df["CUSTOMER_AGE"].count(),[bin_18,bin_30,bin_45,bin_60])})
    
    
```

```python
df_bin["Percentage"] = df["CUSTOMER_AGE"].apply(impute_age_bin,axis=1)
```
